#### Generale_Data_Center_Energia_Attiva.csv

```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import ydata_profiling as ydp
import great_expectations as gx
from datetime import timedelta


dataframe_gdc_ea = pd.read_csv('Generale_Data_Center_Energia_Attiva.csv', sep=',', parse_dates=['data'])

# Report ydata

dataframe_gdc_ea = dataframe_gdc_ea.sort_values(by='data')

type_schema = {"data": "datetime", "energia_attiva": "timeseries"}

report = ydp.ProfileReport(df=dataframe_gdc_ea, title='Generale Data Center Energia Attiva', tsmode=True, sortby="data", type_schema = type_schema)

report.to_file(output_file="Generale_Data_Center_Energia_Attiva.html")

# Report great_expectations

# Carica il DataFrame Pandas
# df = pd.read_csv('Generale_Data_Center_Energia_Attiva.csv', sep=';')

context = gx.get_context()

# Crea un oggetto PandasDataset da un DataFrame Pandas
ge_batch = context.sources.pandas_default.read_csv('Generale_Data_Center_Energia_Attiva.csv')

# Aggiungi aspettative per la tua colonna datetime
ge_batch.expect_column_values_to_be_unique('data')
ge_batch.expect_column_values_to_be_of_type('data', 'datetime64[ns]')  # Assicurati che il tipo sia corretto

# Esegui le aspettative
results = ge_batch.validate()

# Visualizza i risultati
print(results['success'])
```
